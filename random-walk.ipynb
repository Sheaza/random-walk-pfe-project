# Random walk

## Diffusion
If a number of particles subject to Brownian motion are present in a given medium and there is no preferred direction for the random oscillations, then over a period of time the particles will tend to be spread evenly throughout the medium.

```python
# Define parameters for the walk
np.random.seed(420)
dims = 1
T = 10
dt = 0.001
step_n = int(T/dt)
step_set = [-1, 1]
origin = np.zeros((1,dims))

# Simulate steps in 1D
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape)
path = np.concatenate([origin, steps]).cumsum(0)
start = path[:1]
stop = path[-1:]

# Plot the path

fig = plt.figure(figsize=(8,4),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(np.arange(step_n+1), path, c='black',alpha=0.25,s=0.05)
ax.plot(path,c='black',alpha=0.5,lw=0.5,ls='-',)
ax.plot(0, start, c='red', marker='+')
ax.plot(step_n, stop, c='black', marker='o')
plt.title('1D Random Walk')
plt.tight_layout(pad=0)
plt.show()
```

1D single particle movement in time.

```python
# Define parameters for the walk
dims = 2
n_runs = 10
T = 10
dt = 0.01
step_n = int(T/dt)
step_set = [-1, 1]
runs = np.arange(n_runs)
step_shape = (step_n,dims)
# Plot
fig = plt.figure(figsize=(10,10),dpi=250)
ax = fig.add_subplot(111)
ax.set_xlabel('X')
ax.set_ylabel('Y')

for i, col in zip(runs, colors):
    # Simulate steps in 3D
    origin = np.random.randint(low=-10,high=10,size=(1,dims))
    steps = np.random.choice(a=step_set, size=step_shape)
    path = np.concatenate([origin, steps]).cumsum(0)
    start = path[:1]
    stop = path[-1:]
    # Plot the path
    ax.scatter(path[:,0], path[:,1],
                 c=col,alpha=0.15,s=1)
    ax.plot(path[:,0], path[:,1],
              c=col, alpha=0.25,lw=0.25)
    ax.plot(start[:,0], start[:,1],
              c=col, marker='+')
    ax.plot(stop[:,0], stop[:,1],
              c=col, marker='o')

plt.title('2D Random Walk - Multiple molecules')
plt.show()
```

2D multiple particles simultaneous movement. Particles start at points (x,y)  [-10,10] and move in every step (x, y )  {-1, 0, 1}. We can see occurrence of diffusion - density of particles is getting smaller and smaller.

# Comparing distribution of walkers for the diffusion model and random walkers in 1D

```python
M = 10000 # Nr of walkers
L = 100 # Max size of lattice
# Each time step - move walkers and propagate diffusion solution
p = 0.1 # Prob for motion
pinv = 1.0-p
nsteps = 3001 # Nr of timesteps
# Initialize walkers
x = np.zeros(M) # Initial position of walkers
edges = np.array(range(-L,L+1))-0.5
xc = 0.5*(edges[:-1]+edges[1:])
# Initialize concentrations
c = np.zeros((2*L+1,2))
i0 = 0
i1 = 1
c[L] = M # c[L] corresponds to x = 0
cx = range(-L,L+1)
D = p
plt.ion()
noutput = 100
for it in range(nsteps):
    # First update positions of all random walkers
    for iw in range(M):
        rnd = np.random.rand(1)
        dx = -1*(rnd<p)+1*(rnd>pinv)
        x[iw] = x[iw] + dx
        # Perform explicit step for diffusion equation
    for ix in range(1,len(c)-1):
        # use i0 and generate i1
        c[ix,i1] = c[ix,i0] + D*(c[ix-1,i0]-2*c[ix,i0]+c[ix+1,i0])
    # Flip i0 and i1
    ii = i1
    i1 = i0
    i0 = ii
    # Plot the two concentrations
    if it in [100,500,1000,3000]:
        Nx,e = np.histogram(x,edges)
        plt.clf()
        plt.plot(cx,c,'-r',xc,Nx,'-b')
        plt.title(f'iteration: {it}')
        plt.xlabel('x'),plt.ylabel('N'),plt.pause(0.001)
```
